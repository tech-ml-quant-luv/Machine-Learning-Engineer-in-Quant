```python
from dotenv import load_dotenv
import os
import pandas as pd
load_dotenv()
av_api_key = os.getenv("ALPHA_VANTAGE_API_KEY")
```

```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey={av_api_key}'
r = requests.get(url)
data = r.json()
print(data["Time Series (Digital Currency Daily)"])
```

```python
# Convert nested dictionary into DataFrame
df = pd.DataFrame.from_dict(data["Time Series (Digital Currency Daily)"], orient='index')

# Convert index to datetime
df.index = pd.to_datetime(df.index)

# Convert all values to float
df = df.astype(float)

# Sort by date (optional)
df = df.sort_index()

df["4. close"].plot()
```
